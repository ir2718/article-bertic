```python
import numpy as np
import pandas as pd
from rank_bm25 import BM25Okapi
import os
from tqdm import tqdm
from ast import literal_eval
import matplotlib.pyplot as plt
import spacy
```

```python
corpus = pd.read_csv("../data/textual_similarity_lemmas_VERSION_6.csv")
corpus["lemmas"] = corpus["lemmas"].apply(
    lambda x: literal_eval(x)
)

df = pd.read_csv("../data/textual_similarity_recent_no_dups_VERSION_6.csv")

print(df.shape, corpus.shape)
```

```python
SEED = 42
SAMPLE_SIZE = 200
VERSION = 7
TEMPERATURE = 0.8
THRESHOLD_RATIO = 0.7
```

```python
np.random.seed(SEED)

rand_indices_corpus = np.random.randint(0, corpus.shape[0], size=(SAMPLE_SIZE,))
corpus_subset = [c for i, c in enumerate(corpus["lemmas"]) if i in rand_indices_corpus]
corpus_dropped = corpus.drop(rand_indices_corpus.tolist())
corpus_list = corpus_dropped["lemmas"].tolist()
print(len(corpus_subset))

def temperature_scaling_softmax(x, T, threshold_ratio=0):
    mask = x <= x.max() * threshold_ratio
    x = (x - np.mean(x)) / np.std(x)
    x[mask] = float("-inf")
    probs = np.exp(x/T) / (np.exp(x/T)).sum()
    return probs

bm25 = BM25Okapi(corpus_list)
top_sent_lemma_indices = []
for c in tqdm(corpus_subset):
    scores = bm25.get_scores(c)
    scores[scores.argmax()] = 0
    probs = temperature_scaling_softmax(
        scores, T=TEMPERATURE, threshold_ratio=THRESHOLD_RATIO
    )
    index = np.random.choice(np.arange(len(corpus_list)), size=1, replace=False, p=probs)[0]
    top_sent_lemma_indices.append(index)
len(top_sent_lemma_indices)
```

```python
plt.hist(temperature_scaling_softmax(scores, T=TEMPERATURE, threshold_ratio=THRESHOLD_RATIO), bins=100)
plt.show() 
```

```python
pair_indices_in_df = []
for _, i in tqdm(enumerate(top_sent_lemma_indices)):
    pair = corpus_list[i]
    for k, d in corpus.iterrows():
        l = d["lemmas"]
        if l == pair:
            pair_indices_in_df.append(k)
            break
```

```python
df_pair = df.iloc[pair_indices_in_df]
to_remove2 = df_pair.index.tolist()

print(df_pair.shape)
print(len(to_remove2))

df_pair.head()
```

```python
df_final_sample = pd.concat((df_chosen.reset_index(drop=True), df_pair.reset_index(drop=True)), axis=1)
df_final_sample.drop(columns=["features_qids", "qids_set", "features_qids2", "qids_set2"], inplace=True)
df_final_sample.head(15)
```

```python
final_to_remove = to_remove[:]
final_to_remove.extend(to_remove2)
```

```python
df_dropped = df.drop(index=final_to_remove)
print(df_dropped.shape)

nlp = spacy.load("hr_core_news_lg")
new_corpus = []
for i, data in tqdm(df_dropped.iterrows()):
    new_corpus.append([i.lemma_ for i in nlp(data["body"]) if (i not in nlp.Defaults.stop_words) and (not i.is_punct)])
df_new_corpus = pd.DataFrame.from_dict({"lemmas": new_corpus})
```
